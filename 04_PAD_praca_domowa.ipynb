```python
new_categories_tshirt = {'Wh':'White', 'Bl':'Black', 'Tshirt':'T-Shirt', 'T-shirt':'T-Shirt'}
```

```python
def renaming(sentence):
  sentence_table = sentence.split()
  table = []
  for old_word, new_word in new_categories_tshirt.items():
    for word in sentence_table:
      if word == old_word:
        sentence_table[sentence_table.index(word)] = new_word
        
      else:
        continue
  sentence2 = ' '.join(sentence_table)
  return sentence2
```

```python
renaming('Wh Tshirt')
```

Metoda z funkcją zmieniającą

```python
df['tshirt_category'] = df['tshirt_category'].apply(lambda x: renaming(x))
```
